```python
import requests
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("ODDSBLAZE_API_KEY")
if not API_KEY:
    raise ValueError("No API key found. Please set the THE_ODDS_API_KEY environment variable.")

```

```python
url = f'https://arbs.oddsblaze.com/?key={API_KEY}&timezone=America/New_York'
```

```python
r = requests.get(url, timeout=10)
r.raise_for_status()          # raises if HTTP status is 4xx/5xx
data = r.json()
```
